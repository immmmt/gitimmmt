# 打印一页好友信息

```python
import requests
import re
import pandas as pd

session = requests.Session()
session.get('https://tx.com.cn/space/cs/index.do?cf=hd')

users = []
for page in range(1, 34):
    url = f'https://tx.com.cn/im/cs/default.do?page={page}&z=0iOcPKxjMk'
    res = session.get(url)
    # print(res.text)
    # break

    patt = r'friendId=(\d+)&amp;f=1">(.*?)</a>'
    names = re.findall(patt, res.text)
    users.extend(names)
    display(pd.DataFrame(users, columns=['uid', 'name']))
    print(f'第{page}打印完毕')
    break
```

# 打印全部好友信息

```python
import requests
import re
import pandas as pd

session = requests.Session()
session.get('https://tx.com.cn/space/cs/index.do?cf=hd')

users = []
for page in range(1, 34):
    url = f'https://tx.com.cn/im/cs/default.do?page={page}&z=0iOcPKxjMk'
    res = session.get(url)
    # print(res.text)
    # break

    patt = r'friendId=(\d+)&amp;f=1">(.*?)</a>'
    names = re.findall(patt, res.text)
    users.extend(names)
    print(f'第{page}打印完毕')

df = pd.DataFrame(users, columns=['uid', 'name'])
display(df)
```
